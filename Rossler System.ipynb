$\dot{x} = -y-z$\
$\dot{y} = x+ay$\
$\dot{z} = b+z(x-c)$

$a=b=0.2$\
$c=5.7$


$(\textit{c is the variable parameter})$

Importing necessary modules

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

Initial conditions

```python
x0 = 1
y0 = 1
z0 = 1
```

Parameters

```python
a = 0.2
b = 0.2
c = 5.7

t0 = 0
dt = 0.1
```

## Plotting

```python
x_list = []
y_list = []
z_list = []
t_list = []

x1 = x0
y1 = y0
z1 = z0
t1 = t0

for extent in range(100000):
    x_list.append(x1)
    y_list.append(y1)
    z_list.append(z1)
    t_list.append(t1)
    
    x2 = x1 - (y1 + z1)*dt
    y2 = y1 + (x1 + a*y1)*dt
    z2 = z1 + b + z1*(x1 - c)*dt
    t2 = t1 + dt
    
    x1 = x2
    y1 = y2
    z1 = z2
    t1 = t2

plt.figure()
plt.scatter(x_list, y_list, s=0.01, label='xy plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')

plt.figure()
plt.scatter(x_list, z_list, s=0.01, label='xz plot')
plt.xlabel('x')
plt.ylabel('z')
plt.legend(loc='best')

plt.figure()
plt.scatter(y_list, z_list, s=0.01, label='yz plot')
plt.xlabel('y')
plt.ylabel('z')
plt.legend(loc='best')

plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(np.array(x_list), np.array(y_list), np.array(z_list), label="xyz plot", s=0.01) 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.legend(loc='best')
```

## Time-series plots

```python
plt.figure()
plt.scatter(t_list[:1000], x_list[:1000], s=1, label="x versus t")
plt.xlabel("t")
plt.ylabel("x")
plt.legend(loc='upper right')

plt.figure()
plt.scatter(t_list[:1000], y_list[:1000], s=1, label="y versus t")
plt.xlabel("t")
plt.ylabel("y")
plt.legend(loc='upper right')

plt.figure()
plt.scatter(t_list[:1000], z_list[:1000], s=1, label="z versus t")
plt.xlabel("t")
plt.ylabel("z")
plt.legend(loc='upper right')
```

## Lorenz Map for $x$

```python
xmax_list = []
for i in range(1,len(x_list)-1):
    x = x_list[i]
    if (x > x_list[i-1] and x > x_list[i+1]):
        xmax_list.append(x)
plt.scatter(xmax_list[:-1], xmax_list[1:], s=0.1)
plt.xlim(0,14)
plt.ylim(0,14)
```

## Visualizing Chaos in the XY plot

```python
def chaos_plot(c):
    x_list = []
    y_list = []
    z_list = []
    t_list = []

    x1 = x0
    y1 = y0
    z1 = z0
    t1 = t0

    for extent in range(10000):
        x_list.append(x1)
        y_list.append(y1)
        z_list.append(z1)
        t_list.append(t1)

        x2 = x1 - (y1 + z1)*dt
        y2 = y1 + (x1 + a*y1)*dt
        z2 = z1 + b + z1*(x1 - c)*dt
        t2 = t1 + dt

        x1 = x2
        y1 = y2
        z1 = z2
        t1 = t2

    plt.figure()
    plt.scatter(x_list, y_list, s=0.1, label='xy plot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='best')
    plt.xlim(-10,10)
    plt.ylim(-10,10)

    
interact(chaos_plot, c=widgets.FloatSlider(min=1.2,max=10,step=0.1,value=5.7))
```
