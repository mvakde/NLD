```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Time evolution equations: \
$ x(t+1) = y(t) + 1 - ax^2 $\
$ y(t+1) = bx(t) $

$a = 1.4, b=0.3$

Initial Conditions

```python
x0 = 1
y0 = 1
```

Parameters

```python
a = 1.4
b = 0.3

t0 = 0
dt = 1
```

### Plotting the Phase Portrait

```python
x_list = []
y_list = []
t_list = []

x0 = -0.5
y0 = -0.3
t = t0

for t_range in range(10000):
    x_list.append(x0)
    y_list.append(y0)
    t_list.append(t)

    x = y0 + 1 - a*(x0**2)
    y = b*x0
    x0 = x
    y0 = y
    t += dt

plt.scatter(x_list, y_list, c='blue', s=0.01, label='phase portrait')
plt.legend()
```

Interactive Phase Portrait

```python
def phase_portrait(a, b, t0=0, dt=1):

    x_list = []
    y_list = []
    t_list = []

    x0 = -0.5
    y0 = -0.3
    t = t0

    for t_range in range(10000):
        x_list.append(x0)
        y_list.append(y0)
        t_list.append(t)

        x = y0 + 1 - a*(x0**2)
        y = b*x0
        x0 = x
        y0 = y
        t += dt

    plt.scatter(x_list, y_list, c='blue', s=0.01, label='phase portrait')
    plt.legend()
    
interact(phase_portrait, a=widgets.FloatSlider(min=1,max=1.45,step=0.01,value=1.4), b=widgets.FloatSlider(min=0,max=0.4,step=0.01,value=0.3))
```

### Unstable Manifold of the Saddle Point

```python
x_list = []
y_list = []
t_list = []

x0 = -0.5
y0 = -0.3
t = t0

for t_range in range(100000):
    x_list.append(x0)
    y_list.append(y0)
    t_list.append(t)

    x = y0 + 1 - a*(x0**2)
    y = b*x0
    x0 = x
    y0 = y
    t += dt

plt.scatter(x_list, y_list, c='brown', s=0.01, label='Unstable Manifold of the Saddle Point')
plt.legend()
plt.xlim(-1.3,-1)
plt.ylim(0.34,0.4)
```

```python
plt.scatter(x_list, y_list, c='brown', s=0.7, label='Zooming ONCE into the small square')
plt.legend()
plt.xlim(0.4, 0.6)
plt.ylim(0.18, 0.23)
plt.figure()
plt.scatter(x_list, y_list, c='brown', s=0.7, label='Zooming TWICE into the small square')
plt.legend()
plt.xlim(0.475, 0.525)
plt.ylim(0.205, 0.214)
```
